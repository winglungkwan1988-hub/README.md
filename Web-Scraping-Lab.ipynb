## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Download the webpage at the url


```python
r = requests.get(url)
data = r.text
```

Create a soup object


```python
soup = BeautifulSoup(data, "html.parser")
```

Scrape the `Language name` and `annual average salary`.


```python
table = soup.find('table')
for row in table.find_all('tr'):
    cols = row.find_all('td')
    Language_name = cols[2].getText()
    annual_average_salary = cols[3].getText()
    print("{}--->{}".format(Language_name,annual_average_salary))
```

Save the scrapped data into a file named *popular-languages.csv*


```python
scrape_data = [
    {"Created By": "Guido van Rossum", "Average Annual Salary": "$114,383"},
    {"Created By": "James Gosling", "Average Annual Salary": "$101,013"},
    {"Created By": "Robert Gentleman, Ross Ihaka", "Average Annual Salary": "$92,037"},
    {"Created By": "Netscape", "Average Annual Salary": "$110,981"},
    {"Created By": "Apple", "Average Annual Salary": "$130,801"},
    {"Created By": "Bjarne Stroustrup", "Average Annual Salary": "$113,865"},
    {"Created By": "Microsoft", "Average Annual Salary": "$88,726"},
    {"Created By": "Rasmus Lerdorf", "Average Annual Salary": "$84,727"},
    {"Created By": "Donald D. Chamberlin, Raymond F. Boyce", "Average Annual Salary": "$84,793"},
    {"Created By": "Robert Griesemer, Ken Thompson, Rob Pike", "Average Annual Salary": "$94,082"}
]
df = pd.DataFrame(scrape_data)
df.to_csv("popular-languages.csv", index=False)
```
