**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = "SELECT COUNT(*) FROM main"

df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
!pip install seaborn
import seaborn as sns
QUERY = "SELECT ConvertedCompYearly FROM main"
df_comp = pd.read_sql_query(QUERY, conn)

comp_data = df_comp["ConvertedCompYearly"].dropna()

plt.figure(figsize=(10,6))
sns.histplot(comp_data, bins=40, color="skyblue", edgecolor="black")
plt.title("Histogram of Total Compensation (ConvertedCompYearly)")
plt.xlabel("Total Compensation (USD)")
plt.ylabel("Number of Respondents")
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
age_map = {
    "18-24 years": 21,
    "25-34 years": 29,
    "35-44 years": 39,
    "45-54 years": 49,
    "55-64 years": 59,
    "65 years or older": 70
}

QUERY = "SELECT Age FROM main"
df_age = pd.read_sql_query(QUERY, conn)

df_age["AgeNum"] = df_age["Age"].map(age_map)

age_data = df_age["AgeNum"].dropna()
plt.figure(figsize=(8,6))
plt.boxplot(
    age_data,
    patch_artist=True,
    orientation="horizontal",
    boxprops=dict(facecolor="skyblue", color="black"),
    medianprops=dict(color="red")
)
plt.title("Box Plot of Age")
plt.xlabel("Age (Years)")
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
age_map = {
    "18-24 years": 21,
    "25-34 years": 29,
    "35-44 years": 39,
    "45-54 years": 49,
    "55-64 years": 59,
    "65 years or older": 70
}
QUERY = "SELECT Age, WorkExp FROM main"
df_age_exp = pd.read_sql_query(QUERY, conn)

df_age_exp["AgeNum"] = df_age_exp["Age"].map(age_map)

df_clean = df_age_exp.dropna(subset=["AgeNum", "WorkExp"])

plt.figure(figsize=(10,6))
plt.scatter(df_clean["AgeNum"], df_clean["WorkExp"], alpha=0.6, color="steelblue")
plt.title("Scatter Plot of Age vs Work Experience")
plt.xlabel("Age (Years)")
plt.ylabel("Work Experience (Years)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
age_map = {
    "18-24 years": 21,
    "25-34 years": 29,
    "35-44 years": 39,
    "45-54 years": 49,
    "55-64 years": 59,
    "65 years or older": 70
}
QUERY = "SELECT Age, TimeSearching, Frustration FROM main"
df_bubble = pd.read_sql_query(QUERY, conn)

df_bubble["AgeNum"] = df_bubble["Age"].map(age_map)
df_clean = df_bubble.dropna(subset=["AgeNum", "TimeSearching", "Frustration"])

plt.figure(figsize=(10,6))
plt.scatter(df_clean["TimeSearching"], df_clean["Frustration"], s=df_clean["AgeNum"]*2, alpha=0.5, color="skyblue", edgecolor="black")
plt.xlabel("Time Searching")
plt.ylabel("Frustration")
plt.title("Bubble plot: TimeSearching vs Frustration (Bubble Size = Age)")
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
QUERY = "SELECT DatabaseWantToWorkWith FROM main"
df_db = pd.read_sql_query(QUERY, conn)

db_counts = df_db["DatabaseWantToWorkWith"].value_counts()
top5 = db_counts.head(5)
plt.figure(figsize=(10,6))
plt.pie(top5.values, labels=top5.index, autopct="%1.1f%%")
plt.title("Pie Chart of top5: wish to learn next year")
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
age_map = {
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years old or older": 70
}
QUERY = "SELECT Age, TimeSearching, TimeAnswering FROM main"
df_median = pd.read_sql_query(QUERY, conn)

df_median["AgeNum"] = df_median["Age"].map(age_map)

df_30_35 = df_median[df_median["Age"] == "25-34 years old"]

df_30_35["TimeSearching"] = pd.to_numeric(df_30_35["TimeSearching"], errors="coerce")
df_30_35["TimeAnswering"] = pd.to_numeric(df_30_35["TimeAnswering"], errors="coerce")

median_searching = df_30_35["TimeSearching"].median(skipna=True)
median_answering = df_30_35["TimeAnswering"].median(skipna=True)

print("Median Searching:", median_searching)
print("Median Answering:", median_answering)

plt.figure(figsize=(8,6))
plt.bar("Age 30-35", median_searching, label="Median Time Searching", color="skyblue")
plt.bar("Age 30-35", median_answering, bottom=median_searching, label="Median Time Answering", color="salmon")

plt.title("Stacked Bar Chart of Median TimeSearching and TimeAnswering (Age 30-35)")
plt.ylabel("Median Time (Hours)")
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
age_map = {
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years old or older": 70
}
QUERY = "SELECT Age, ConvertedCompYearly FROM main"
df_age = pd.read_sql_query(QUERY, conn)

df_age["AgeNum"] = df_age["Age"].map(age_map)

df_age["ConvertedCompYearly"] = pd.to_numeric(df_age["ConvertedCompYearly"], errors="coerce")

df_target = df_age[(df_age["AgeNum"] >= 45) & (df_age["AgeNum"] <= 60)]
median_age = df_target.groupby("Age")["ConvertedCompYearly"].median()
plt.figure(figsize=(8,6))
plt.plot(median_age.index, median_age.values, marker=0, color="skyblue")
plt.xlabel("Age Group")
plt.ylabel("Median Compendation(USD)")
plt.title("Line Chart of Median Compensation from 45 to 60")
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
QUERY = "SELECT MainBranch FROM main"
df_branch = pd.read_sql_query(QUERY, conn)

branch_counts = df_branch["MainBranch"].value_counts()

plt.figure(figsize=(8,6))
plt.barh(branch_counts.index, branch_counts.values, color="blue")
plt.xlabel("Count")
plt.ylabel("MainBranch")
plt.title("Distribution of Main Branch")
plt.show()
```
